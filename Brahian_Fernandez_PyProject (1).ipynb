```python
path = '/content/drive/MyDrive/ecommerce_orders.csv'
df = pd.read_csv(path)
quantity = df['Quantity'].to_numpy()
print("Mean:",  np.mean(quantity))
print("Median:", np.median(quantity))
```

####2. Use NumPy to calculate the maximum and minimum order Price.

```python
price = df['Price'].to_numpy()
print("Max:", np.max(price))
print("Min:", np.min(price))
```

####3. Create a NumPy array of Quantity * Price (i.e., total revenue per order), print first 10 values(Hint: use slice operator)

```python
revperorder = quantity * price
print("Total revenue for first 10 orders:", revperorder[:10])
```

####4. Normalize the Quantity values using NumPy (scaling between 0 and 1).

```python
normalizedquants = (quantity - np.min(quantity))/(np.max(quantity) - np.min(quantity))
print("Normalized quantity:", normalizedquants)
```

####20. Create a pie chart for product sales distribution.

```python
productsales = df.groupby('Product')['TotalAmount'].sum()
plt.figure(figsize=(10, 10))
plt.pie(productsales, labels = productsales.index, autopct='%1.1f%%')
plt.title("Product Sales Distribution")
plt.show()
```
